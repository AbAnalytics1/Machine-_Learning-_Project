### Dropping columns with lots of missing values. 

For loans that are late, they cannot be classified as defaulted because although the repayment is just late and not defaulted. A loan becomes defaulted when the borrower fails to make payment 60 days after the grace period (according to https://ccr.equifax.com.au/resources/explaining-the-difference-between-a-late-payment-and-a-default-to-your-customers#:~:text=Repayment%20history%20information%20is%20recorded,was%20in%20a%20specific%20month.&text=A%20payment%20later%20than%2060%20days%20is%20a%20default.) Since this date can only be defined by the loan company, this might be difficult to achieve. However, the loan company has provided us with a Defaultdate column which is the date the loan went into defaulted state. 

This means for rows or borrowers that have a default date, the borrower is assumed to have defaulted the payment of the loan. This means that the default date is what will be used to determine whether a borrower has defaulted a loan or not. This also means that the defaultdate column cannot be dropped. 

```python
df = df.drop(columns=['NrOfDependants', 'WorkExperience', 'CreditScoreEsMicroL', 
'CreditScoreEsEquifaxRisk', 'CreditScoreFiAsiakasTietoRiskGrade', 'CreditScoreEeMini', 
'PreviousRepaymentsBeforeLoan', 'PreviousEarlyRepaymentsBefoleLoan'])
df.info()
```

### Categorical vs Continous variables
I will need to convert all the categoriacl variables to features. Looking at the datasets, data like gender, verificationtype, newcreditcustomer etc, are categorical variables as they have meanings beyond being just numbers. Income total however is a value and not a category of values. 

I will be using one hot encoding instead of label encoding to convert all the categorical variables into numerical variable. Although label encoding is more direct and easy to implement, it can throw off the model as it assigns values to numbers based on their alphabetical order, the model can therefore assume that 4 is greater than 1 and make a prediction based on that.

One hot encoding however, creates additional features based on the number of unique values in the categorical feature. It has the disadvantage of adding more columns to the dataset. Every unique value in the category will be added as a feature.

Identifying the categorical variables.

The categorical variables are: 

<li>Gender</li>
<li>VerificationType </li>
<li>NewCreditCustomer</li>
<li>UseOfLoan</li>
<li>Country </li>
<li>LoanDuration </li>
<li>Education</li>
<li>MaritalStatus</li>
<li>EmploymentStatus</li>
<li>EmploymentDurationCurrentEmployer</li>
<li>OccupationArea</li>
<li>HomeOwnershipType</li>

The default column will be left alone since it only involves two variables 1 and 0. Similar to what's expected from label encoding. Using label encoding this way is fine. 

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder()

df = df.reset_index(drop=True)
#Passing encoded columns
col_names = ['Gender', 'VerificationType', 'NewCreditCustomer', 'UseOfLoan', 'Country', 
'LoanDuration', 'Education', 'MaritalStatus', 'EmploymentStatus', 'OccupationArea', 'HomeOwnershipType', 'EmploymentDurationCurrentEmployer']

enc_data =pd.DataFrame(enc.fit_transform(df[['Gender', 'VerificationType', 'NewCreditCustomer', 'UseOfLoan', 'Country', 
'LoanDuration', 'Education', 'MaritalStatus', 'EmploymentStatus', 'OccupationArea', 'HomeOwnershipType', 'EmploymentDurationCurrentEmployer']]).toarray(), columns=enc.get_feature_names(col_names))

#Merge with main
res_df = df.join(enc_data)
res_df.drop(columns= ['Gender', 'VerificationType', 'NewCreditCustomer', 'UseOfLoan', 'Country', 
'LoanDuration', 'Education', 'MaritalStatus', 'EmploymentStatus', 'OccupationArea', 
'HomeOwnershipType', 'EmploymentDurationCurrentEmployer'], inplace = True)
res_df.head()
```

Splitting the entire dataset to test and train data.

```python
# Extracting the dependent and independent variables. ALl the variables are independent except the default column which we derived earlier.

y = res_df['Default']
X = res_df.loc[:, res_df.columns != 'Default']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

## Random Forest

```python
# Building the model

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error


model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=0).fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(predictions, y_test)
print('Accuracy Score is', accuracy)

rf_val_mae = mean_absolute_error(predictions, y_test)

print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```

#### Getting the best n_estimator to use for the randomforest model

```python
def get_mae(estimator, X_train, X_test, y_train, y_test):
    model = RandomForestClassifier(n_estimators=estimator, max_depth=10, random_state=0).fit(X_train, y_train)
    predictions = model.predict(X_test)
    mae = mean_absolute_error(predictions, y_test)
    return(mae)
```

```python
candidate_estimators = [100, 200, 300, 400, 500, 1000]

values = {estimator :  get_mae(estimator, X_train, X_test, y_train, y_test)
          for estimator in candidate_estimators}

# print(values)

best_estimator = min(values, key = values.get)
print("The best n_estimator values is", best_estimator)
```

#### Getting the best depth to use for the random forest model

```python
def get_mae(depth, X_train, X_test, y_train, y_test):
    model = RandomForestClassifier(n_estimators=100, max_depth=depth, random_state=0).fit(X_train, y_train)
    predictions = model.predict(X_test)
    mae = mean_absolute_error(predictions, y_test)
    return(mae)
```

```python
candidate_depths = [1, 2, 4, 6, 8, 10, 12, 14, 20, 50, 100]
right_depth = {depth: get_mae(depth, X_train, X_test, y_train, y_test) for depth in candidate_depths}
best_depth = min(right_depth, key = right_depth.get)
print("The best depth is", best_depth)
```

## Final Model selction for random forest.

Based on the provided parameters above, I will select the parameters that are regarded as the best parameters to use and then use them on the actual model below. 

```python
model = RandomForestClassifier(n_estimators=400, max_depth=100, random_state=0).fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(predictions, y_test)
print('Accuracy Score is', accuracy)

rf_val_mae = mean_absolute_error(predictions, y_test)

print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```

### Decision Tree
Guide for selecting parameters: https://towardsdatascience.com/how-to-tune-a-decision-tree-f03721801680

```python
def get_mae_dtc(depth, X_train, X_test, y_train, y_test):
    clf = DecisionTreeClassifier(max_depth = depth, random_state = 0).fit(X_train, y_train)
    predictor = clf.predict(X_test)
    rf_val_mae = mean_absolute_error(predictor, y_test)
    return(rf_val_mae)
```

```python
candidate_depths = [1, 2, 4, 6, 8, 10, 12, 13, 14, 20, 50, 100]
right_depth = {depth: get_mae_dtc(depth, X_train, X_test, y_train, y_test) for depth in candidate_depths}
print(right_depth)
best_depth = min(right_depth, key = right_depth.get)
print("The best depth is", best_depth)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=200).fit(X_train, y_train)
predictor = model.predict(X_test)
accuracy = accuracy_score(predictor, y_test)
print('Accuracy Score is', accuracy)
```

```python
def get_mae_dtc(neighbor, X_train, X_test, y_train, y_test):
    clf = KNeighborsClassifier(n_neighbors=neighbor).fit(X_train, y_train)
    predictor = clf.predict(X_test)
    rf_val_mae = mean_absolute_error(predictor, y_test)
    return(rf_val_mae)
```

```python
candidate_neighbors = np.arange(1, 500, 10)
right_neighbor = {depth: get_mae_dtc(depth, X_train, X_test, y_train, y_test) for depth in candidate_neighbors}
print(right_neighbor)
best_neighbor = min(right_neighbor, key = right_neighbor.get)
print("the best neighbor value is", best_neighbor)
```
