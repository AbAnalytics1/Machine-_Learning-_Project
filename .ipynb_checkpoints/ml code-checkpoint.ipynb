Sales type, show and xero prod will not feature in the correlation yet as they are still strings. I will be using one hot enccoding for them instead of label encoding. My reasons for this is detailed in the coming sections. 

```python
df.corr()
```

I will be using one hot encoding instead of label encoding to convert the categorical variables into numerical variable. ALthough label encoding is more direct and easy to implement, it can throw off the model as it assigns values to numbers based on their alphabetical order, the model can therefore assume that 4 is greater than 1 and make a prediction based on that. 

One hot encoding however, creates additional features based on the number of unique values in the categorical feature. It has the disadvantage of adding more columns to the dataset.Every unique value in the category will be added as a feature.

Before performing the encoding, I realised that it is advisable for the encoding on the training dataset and that of the test dataset should be different. The explanation to that is detailed below. 

Generally, you want to treat the test set as though you did not have it during training. Whatever transformations you do to the train set should be done to the test set before you make predictions. So yes, you should do the transformation separately, but know that you are applying the same transformation.

For example, if the test set is missing one of the categories, there should still be a dummy variable for the missing category (which would be found in the training set), since the model you train will still expect that dummy variable. If the test set has an extra category, this should probably be handled with some "other" category.

Link: https://stackoverflow.com/questions/55525195/do-i-have-to-do-one-hot-encoding-separately-for-train-and-test-dataset

```python
from sklearn.preprocessing import OneHotEncoder
enc=OneHotEncoder()

df = df.reset_index(drop=True)
#Passing encoded columns
col_names = ['SHOW','SALES TYPE', 'XERO PROD', 'ACCOUNT MANAGER']

enc_data =pd.DataFrame(enc.fit_transform(df[['SHOW','SALES TYPE', 'XERO PROD', 'ACCOUNT MANAGER']]).toarray(), columns=enc.get_feature_names(col_names))

#Merge with main
res_df = df.join(enc_data)
res_df.drop(columns= ['SHOW','SALES TYPE', 'XERO PROD', 'ACCOUNT MANAGER'], inplace = True)
res_df.head()
```

 ### Random Forest

```python
# Building the model

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error


model = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=0).fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(predictions, y_test)
print('Accuracy Score is', accuracy)

rf_val_mae = mean_absolute_error(predictions, y_test)

print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```

Get the best n_estimator to use for the randomforest model

```python
def get_mae(estimator, X_train, X_test, y_train, y_test):
    model = RandomForestClassifier(n_estimators=estimator, max_depth=10, random_state=0).fit(X_train, y_train)
    predictions = model.predict(X_test)
    mae = mean_absolute_error(predictions, y_test)
    return(mae)
```

```python
candidate_estimators = [100, 200, 300, 400, 500, 1000]

values = {estimator :  get_mae(estimator, X_train, X_test, y_train, y_test)
          for estimator in candidate_estimators}

print(values)

best_estimator = min(values, key = values.get)
print(best_estimator)
```

Getting the best depth to use for the random forest model

```python
def get_mae(depth, X_train, X_test, y_train, y_test):
    model = RandomForestClassifier(n_estimators=100, max_depth=depth, random_state=0).fit(X_train, y_train)
    predictions = model.predict(X_test)
    mae = mean_absolute_error(predictions, y_test)
    return(mae)
```

```python
candidate_depths = [1, 2, 4, 6, 8, 10, 12, 14, 20, 50, 100]
right_depth = {depth: get_mae(depth, X_train, X_test, y_train, y_test) for depth in candidate_depths}
print(right_depth)
best_depth = min(right_depth, key = right_depth.get)
print(best_depth)
```

In selecting the best estimator and depth, I will use 100 and 50. 100 because with 1000, the model appears to be very slow, while 100, 200 and 300 have a higher mean square error than 100. I will use 50 for the depth because it has the lowest mean square error.

```python
%%time
# Using the best randomforestclassifier model parameters

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import mean_absolute_error


model = RandomForestClassifier(n_estimators=100, max_depth=50, random_state=0).fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy = accuracy_score(predictions, y_test)
print('Accuracy Score is', accuracy)

rf_val_mae = mean_absolute_error(predictions, y_test)

print("Validation MAE for Random Forest Model: {}".format(rf_val_mae))
```

### Decision Tree 

`Guide for selecting parameters:` https://towardsdatascience.com/how-to-tune-a-decision-tree-f03721801680

I will be using the same random state that I used in random forest in order for me to work with the same data randomness in both models

```python
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(random_state = 0).fit(X_train, y_train)
predictor = clf.predict(X_test)
accuracy = accuracy_score(predictor, y_test)
print('Accuracy Score is', accuracy)


rf_val_mae = mean_absolute_error(predictor, y_test)

print("Validation MAE for Decision Tree Model: {}".format(rf_val_mae))
```

```python
def get_mae_dtc(depth, X_train, X_test, y_train, y_test):
    clf = DecisionTreeClassifier(max_depth = depth, random_state = 0).fit(X_train, y_train)
    predictor = clf.predict(X_test)
    rf_val_mae = mean_absolute_error(predictor, y_test)
    return(rf_val_mae)
```

```python
candidate_depths = [1, 2, 4, 6, 8, 10, 12, 13, 14, 20, 50, 100]
right_depth = {depth: get_mae_dtc(depth, X_train, X_test, y_train, y_test) for depth in candidate_depths}
print(right_depth)
best_depth = min(right_depth, key = right_depth.get)
print(best_depth)
```

```python
%%time
from sklearn.tree import DecisionTreeClassifier
clf = DecisionTreeClassifier(max_depth = 13, random_state = 0).fit(X_train, y_train)
predictor = clf.predict(X_test)
accuracy = accuracy_score(predictor, y_test)
print('Accuracy Score is', accuracy)


rf_val_mae = mean_absolute_error(predictor, y_test)

print("Validation MAE for K Decision Tree Model: {}".format(rf_val_mae))
```

```python
def get_mae_dtc(neighbor, X_train, X_test, y_train, y_test):
    clf = KNeighborsClassifier(n_neighbors=neighbor).fit(X_train, y_train)
    predictor = clf.predict(X_test)
    rf_val_mae = mean_absolute_error(predictor, y_test)
    return(rf_val_mae)
```

```python
candidate_neighbors = np.arange(1, 500, 10)
right_neighbor = {depth: get_mae_dtc(depth, X_train, X_test, y_train, y_test) for depth in candidate_neighbors}
print(right_neighbor)
best_neighbor = min(right_neighbor, key = right_neighbor.get)
print(best_neighbor)
```
